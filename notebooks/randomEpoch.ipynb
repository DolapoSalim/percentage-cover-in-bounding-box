```python
import numpy as np
import matplotlib.pyplot as plt

# Simulate data for 5 runs, 50 epochs each
np.random.seed(42)
epochs = np.arange(1, 51)
runs = 5

# Generate synthetic accuracy data (mean accuracy increases over time, with some noise)
all_runs = [0.5 + 0.01 * epochs + np.random.normal(0, 0.02, size=epochs.size) for _ in range(runs)]
all_runs = np.array(all_runs)

# Calculate mean and standard deviation across runs
mean_accuracy = all_runs.mean(axis=0)
std_accuracy = all_runs.std(axis=0)

# Plot mean learning curve
plt.plot(epochs, mean_accuracy, label="Mean Accuracy", color="blue")

# Add shaded area for 1 standard deviation
plt.fill_between(epochs, mean_accuracy - std_accuracy, mean_accuracy + std_accuracy,
                 color="blue", alpha=0.2, label="1 Std Dev")

# Optional: plot individual runs
for i in range(runs):
    plt.plot(epochs, all_runs[i], color="gray", alpha=0.3, linestyle="--")

# Formatting
plt.title("Model Accuracy Over Epochs with Variability")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
